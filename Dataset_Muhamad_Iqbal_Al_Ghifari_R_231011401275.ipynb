```python
hasilgabung =pd.concat([ar,ir,ur,er,ro],ignore_index=True)
```

```python
hasilgabung.columns.values
```

```python
A =  hasilgabung.iloc[:,7:76]
```

```python
A
```

```python
B = hasilgabung.iloc[:,83:84]
```

```python
B
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
A_train,A_test,B_train,B_test = train_test_split(A,B, test_size = 0.2, random_state = 42)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
santi = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
santi.fit(A_train,B_train)
```

```python
B_pred = santi.predict(A_test)
```

```python
B_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(B_test,B_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(santi, feature_names = A.columns.values, class_names = np.array([ 'DDoS UDP Flood','DoS ICMP Flood','MITM ARP Spoofing', 'MQTT DoS Publish Flood', 'Recon Vulnerability Scan']), filled = True)
plt.show()
```
